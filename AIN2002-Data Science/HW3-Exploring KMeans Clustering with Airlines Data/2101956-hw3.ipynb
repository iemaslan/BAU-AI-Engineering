```python
import pandas as pd

# Load the dataset
data = pd.read_csv('/content/airlines.csv')

# Display the first few rows of the dataset
print(data.head())

# Display summary statistics
print(data.describe())

# Display information about the dataset
print(data.info())
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Option 1: Fill missing values
data.fillna(data.mean(), inplace=True)

# Option 2: Drop missing values
data.dropna(inplace=True)
```

```python
import matplotlib.pyplot as plt

# Check the column names
print(data.columns)

# List of features to plot
features = ['balance', 'qual_mile', 'cc1_miles', 'cc2_miles', 'cc3_miles', 'bonus_miles', 'bonus_trans', 'flight_miles_12mo', 'flight_trans_12', 'days_since_enrolled']

# Ensure the feature list is correct
features = [feature for feature in features if feature in data.columns]

# Plot histograms for each feature
plt.figure(figsize=(15, 15))
for i, feature in enumerate(features, 1):
    plt.subplot(5, 2, i)
    plt.hist(data[feature], bins=15, edgecolor='k')
    plt.title(feature)
    plt.xlabel(feature)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

```

```python
# Calculate the percentage of customers with and without award flights
award_counts = data['award'].value_counts(normalize=True) * 100
print(f"Percentage of customers without an award flight: {award_counts[0]:.2f}%")
print(f"Percentage of customers with an award flight: {award_counts[1]:.2f}%")
```

```python
import seaborn as sns

# Calculate correlation matrix
corr_matrix = data.corr()

# Display the correlation matrix
print(corr_matrix)

# Plot a heatmap of the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()

# Find features most correlated with 'balance'
print(corr_matrix['balance'].sort_values(ascending=False))
```

```python
# Plot the relationship between 'bonus_miles' and 'bonus_trans'
plt.figure(figsize=(8, 6))
plt.scatter(data['bonus_miles'], data['bonus_trans'], alpha=0.5)
plt.xlabel('Bonus Miles')
plt.ylabel('Bonus Transactions')
plt.title('Relationship between Bonus Miles and Bonus Transactions')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Normalize the data
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data.drop(columns=['id', 'award']))

# Determine the optimal number of clusters using the Elbow Method
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)
    sse.append(kmeans.inertia_)

# Plot the SSE for each k
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('Elbow Method for Optimal k')
plt.show()

# Determine the optimal number of clusters using Silhouette Score
silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)
    score = silhouette_score(data_scaled, kmeans.labels_)
    silhouette_scores.append(score)

# Plot the Silhouette Scores for each k
plt.figure(figsize=(8, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal k')
plt.show()
```

```python
# Fit the KMeans algorithm with the chosen number of clusters
optimal_k = 3  # Replace with the chosen k based on the plots
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(data_scaled)

# Add the cluster labels to the original dataset
data['cluster'] = clusters

# Analyze the clusters
print(data.groupby('cluster').mean())
```
